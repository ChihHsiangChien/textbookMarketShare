```python
from pandas import DataFrame
import pandas as pd

df = pd.read_csv('教科書版本資料.csv')
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mapclassify
```

```python
shpFile = '直轄市、縣市界線(TWD97經緯度)/COUNTY_MOI_1090820.shp'
gdf = gpd.read_file(shpFile)
```

```python
#根據某一科儲存單一科的檔案
filter1 = df['學制'] == '國中'
filter2 = df['科目'] == '社會'
filter3 = df['年級'] > 6  

df_nature = df[filter1  & filter2 & filter3  ]

table = pd.pivot_table(df_nature,
                       values='學校', 
                       #index=['縣市', '區域'],
                       index=['縣市'],
                       
                       columns='版本',

                       aggfunc='count',
                       fill_value=0)
#換算成比例
cityTable = table.apply(lambda r: r/r.sum(), axis=1)
cityTable
```

```python
#pd.merge(gdf, cityTable, left_on='COUNTYNAME',).drop('縣市',axis=1)
gdf_new = pd.merge(gdf, cityTable, left_on='COUNTYNAME',right_on=cityTable.index)
```

```python
#找出此張地圖的邊界
minx, miny, maxx, maxy = gdf_new.total_bounds

print(minx, miny, maxx, maxy)
```

```python
#設定此張地圖的邊界

fig, axes = plt.subplots(figsize=(5, 3), dpi=200, nrows=1, ncols=3, )
fig.suptitle('教科書版圖',fontproperties = myfont)

vmax = 1.0
vmin = 0

for ax in axes:
    ax.set_xlim(120, 122)
    ax.set_ylim(21.5, 25.5)
    im = gdf_new.plot(ax = ax ,
                 column = '南一',
                 cmap = 'Blues',
                 linewidth = 0.2,
                 edgecolor = 'gray',
                 #legend = True,
                 vmax = vmax,
                 vmin = vmin)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])
    ax.set_title('               '+ '南一', size=10,fontproperties = myfont)
    
    for index, row in gdf_new.iterrows():
        xy = row['geometry'].centroid.coords[:]
        xytext = row['geometry'].centroid.coords[:]
        ax.annotate(row['COUNTYNAME'], xy=xy[0], xytext=xytext[0],
                     horizontalalignment='center', verticalalignment='center',fontproperties = myfont,size=3,)    
    

#c=畫出 colorbar    
cpos = [axes[0].get_position().bounds[1],
        axes[0].get_position().bounds[3]]
cax = fig.add_axes([0.95,cpos[0],0.01,cpos[1]])
fig.subplots_adjust(right=0.8)
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm,cax = cax)


fig.savefig('test.png')

```

# 設計迴圈產生圖片

```python
from matplotlib.font_manager import *
#myfont = FontProperties(fname=r'C:\Windows\Fonts\NotoSansCJKtc-Regular.otf')
myfont = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')
#plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def createMap(gdf,schoolType,subject):
    
    #設定此張地圖的邊界
    fig, axes = plt.subplots(figsize=(5, 3), dpi=200, nrows=1, ncols=3, )
    fig.suptitle(schoolType + '  '+ subject + '教科書版圖',fontproperties = myfont)

    vmax = 1.0
    vmin = 0
    versions = ['南一','翰林','康軒']
    i = 0
    for ax in axes:
        ax.set_xlim(120, 122)
        ax.set_ylim(21.5, 25.5)
        im = gdf.plot(ax = ax ,
                     column = versions[i],
                     cmap = 'Blues',
                     linewidth = 0.2,
                     edgecolor = 'gray',
                     #legend = True,
                     vmax = vmax,
                     vmin = vmin)

        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
        ax.set_title('               '+ versions[i] , size=10,fontproperties = myfont)
        
        for index, row in gdf.iterrows():
            xy = row['geometry'].centroid.coords[:]
            xytext = row['geometry'].centroid.coords[:]
            ax.annotate(row['COUNTYNAME'], xy=xy[0], xytext=xytext[0],
                         horizontalalignment='center', verticalalignment='center',fontproperties = myfont,size=3,)         
                     
        i += 1


    #c=畫出 colorbar    
    cpos = [axes[0].get_position().bounds[1],
            axes[0].get_position().bounds[3]]
    cax = fig.add_axes([0.9,cpos[0],0.01,cpos[1]])
    fig.subplots_adjust(right=0.8)
    sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm._A = []
    cbar = fig.colorbar(sm,cax = cax)


    fig.savefig('分析圖/' + schoolType + '_'+ subject + '.png')
```

```python
#根據某一科儲存單一科的檔案

for schoolType in ['國小','國中']:
    filter1 = df['學制'] == schoolType
    subjects = df[filter1]['科目'].unique()
    for subject in subjects:
        filter2 = df['科目'] == subject
        df_temp = df[filter1  & filter2 ]
        
        
        table = pd.pivot_table(df_temp,
                       values='學校', 
                       #index=['縣市', '區域'],
                       index=['縣市'],
                       columns='版本',
                       aggfunc='count',
                       fill_value=0)
        #換算成比例
        cityTable = table.apply(lambda r: r/r.sum(), axis=1)

        
        gdf_new = pd.merge(gdf, cityTable, left_on='COUNTYNAME',right_on=cityTable.index)
        
        createMap(gdf_new,schoolType,subject)
```
